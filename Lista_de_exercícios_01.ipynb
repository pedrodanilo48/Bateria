### Exerccio 09
Escreva um programa que leia dois nmeros e que pergunte qual operao voc deseja realizar. Voc deve poder calcular a soma (+), subtrao (-), multiplicao (*) e diviso (/). Exiba o resultado da operao solicitada.



```python
primeironumero = float(input("Digite o primeiro nmero: "))
segundonumero = float(input("Digite o segundo nmero: "))
operacao = input("Digite a operao que deseja realizar: ")
if operacao == "+":
  print(primeironumero+segundonumero)
  if operacao == "-":
    print(primeironumero-segundonumero)
    if operacao == "*":
      print(primeironumero*segundonumero)
      if operacao == "/":
        print(primeironumero/segundonumero)
```

### Exerccio 10
Escreva um programa para aprovar o emprstimo bancrio para compra de uma casa. O programa deve receber o valor da casa a ser comprada, o salrio do comprador e a quantidade de anos a pagar. O valor da prestao mensal no pode ser superior a 30% do salrio. Calcule o valor da prestao como sendo o valor da casa a comprar dividido pelo nmero de meses a pagar. O programa deve apresentar as mensagens "Emprstimo aprovado" ou "Emprstimo no aprovado".



```python
casa = float(input("Digite o valor da casa: "))
salario = float(input("Digite o salrio do comprador: "))
anos = int(input("Digite a quantidade de anos a pagar"))
meses = anos*12
prestacao = casa/meses
if prestacao > 0.3*salario:
  print("Emprstimo no aprovado")
else:
  print("Emprstimo aprovado")
```
